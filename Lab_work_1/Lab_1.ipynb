```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Функции для численного вычисления производной

def practical_derivative_sin_sqr_x_1(x, h):
    return (np.sin((x + h) ** 2) - np.sin(x ** 2)) / h

def practical_derivative_sin_sqr_x_2(x, h):
    return (np.sin((x) ** 2) - np.sin((x - h) ** 2)) / h

def practical_derivative_sin_sqr_x_3(x, h):
    return (np.sin((x + h) ** 2) - np.sin((x - h) ** 2)) / (2 * h)

def practical_derivative_sin_sqr_x_4(x, h):
    return 4 / 3 * (np.sin((x + h) ** 2) - np.sin((x - h) ** 2)) / (2 * h) - 1 / 3 * (np.sin((x + 2*h) ** 2) - np.sin((x - 2*h) ** 2)) / (4 * h)

def practical_derivative_sin_sqr_x_5(x, h):
    return 3 / 2 * (np.sin((x + h) ** 2) - np.sin((x - h) ** 2)) / (2 * h) - 3 / 5 * (np.sin((x + 2*h) ** 2) - np.sin((x - 2*h) ** 2)) / (4 * h) + 1 / 10 * (np.sin((x + 3*h) ** 2) - np.sin((x - 3*h) ** 2)) / (6 * h)

# Функция для теоретического расчёта производной
def theoretical_derivative_sin_sqr_x(x):
    return 2 * x * np.cos(x ** 2)
```

```python
# Первая функция y = sin(x^2)
# y' = 2x*cos(x^2)

# Точка, в которой рассматриваются производные
x_0 = 2

# Количество шагов для h
max_n = 21

# Шаг численного дифференцирования
h = []

# Массивы, в которые будет записан результат разности 
f_1 = []
f_2 = []
f_3 = []
f_4 = []
f_5 = []

# Теоретическое значение производной в точке
f_deriv_th = theoretical_derivative_sin_sqr_x(x_0)

# В цикле заполняем массивы модулями разности значений теоретически и практически рассчитанной производной (5 способами)
for i in range(max_n):
    h.append(1 / 2 ** i)
    f_1.append(abs(f_deriv_th - practical_derivative_sin_sqr_x_1(x_0, h[i])))
    f_2.append(abs(f_deriv_th - practical_derivative_sin_sqr_x_2(x_0, h[i])))
    f_3.append(abs(f_deriv_th - practical_derivative_sin_sqr_x_3(x_0, h[i])))
    f_4.append(abs(f_deriv_th - practical_derivative_sin_sqr_x_4(x_0, h[i])))
    f_5.append(abs(f_deriv_th - practical_derivative_sin_sqr_x_5(x_0, h[i])))

# Строим графики ln(df') = F(ln(h))
plt.figure(figsize=[12, 4])
plt.loglog(h, f_1, label='1st')
plt.loglog(h, f_2, label='2nd')
plt.loglog(h, f_3, label='3rd')
plt.loglog(h, f_4, label='4th')
plt.loglog(h, f_5, label='5th')

plt.title("Error rate sin(x^2)")
plt.xlabel("$log(h)$")
plt.ylabel("$log(df')$")
plt.grid()
plt.legend(loc='best', bbox_to_anchor=(1, 1))
plt.tight_layout() 
plt.show()
```

```python
# Упрощённое для записи выражение функции, для которой производится анализ
def c_s(x):
    return np.cos(np.sin(x))

# Функции для численного вычисления производной

def practical_derivative_cos_sin_x_1(x, h):
    return (c_s(x + h) - c_s(x)) / h

def practical_derivative_cos_sin_x_2(x, h):
    return (c_s(x) - c_s(x - h)) / h

def practical_derivative_cos_sin_x_3(x, h):
    return (c_s(x + h) - c_s(x - h)) / (2 * h)

def practical_derivative_cos_sin_x_4(x, h):
    return 4 / 3 * (c_s(x + h) - c_s(x - h)) / (2 * h) - 1 / 3 * (c_s(x + 2*h) - c_s(x - 2*h)) / (4 * h)

def practical_derivative_cos_sin_x_5(x, h):
    return 3 / 2 * (c_s(x + h) - c_s(x - h)) / (2 * h) - 3 / 5 * (c_s(x + 2*h) - c_s(x - 2*h)) / (4 * h) + 1 / 10 * (c_s(x + 3*h) - c_s(x - 3*h)) / (6 * h)

# Функция для теоретического расчёта производной
def theoretical_derivative_cos_sin_x(x):
    return -np.sin(np.sin(x)) * np.cos(x)
```

```python
# Вторая функция y = cos(sin(x))
# y' = -sin(sin(x))*cos(x)

# Точка, в которой рассматриваются производные
x_0 = 2

# Количество шагов для h
max_n = 21

# Шаг численного дифференцирования
h = []

# Массивы, в которые будет записан результат разности 
f_1 = []
f_2 = []
f_3 = []
f_4 = []
f_5 = []

# Теоретическое значение производной в точке
f_deriv_th = theoretical_derivative_cos_sin_x(x_0)

# В цикле заполняем массивы модулями разности значений теоретически и практически рассчитанной производной (5 способами)
for i in range(max_n):
    h.append(1 / 2 ** i)
    f_1.append(abs(f_deriv_th - practical_derivative_cos_sin_x_1(x_0, h[i])))
    f_2.append(abs(f_deriv_th - practical_derivative_cos_sin_x_2(x_0, h[i])))
    f_3.append(abs(f_deriv_th - practical_derivative_cos_sin_x_3(x_0, h[i])))
    f_4.append(abs(f_deriv_th - practical_derivative_cos_sin_x_4(x_0, h[i])))
    f_5.append(abs(f_deriv_th - practical_derivative_cos_sin_x_5(x_0, h[i])))

# Строим графики ln(df') = F(ln(h))
plt.figure(figsize=[12, 4])
plt.loglog(h, f_1, label='1st')
plt.loglog(h, f_2, label='2nd')
plt.loglog(h, f_3, label='3rd')
plt.loglog(h, f_4, label='4th')
plt.loglog(h, f_5, label='5th')

plt.title("Error rate cos(sin(x))")
plt.xlabel("$log(h)$")
plt.ylabel("$log(df')$")
plt.grid()
plt.legend(loc='best', bbox_to_anchor=(1, 1))
plt.tight_layout() 
plt.show()
```
